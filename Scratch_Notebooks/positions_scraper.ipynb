```python
import requests
import pandas as pd
from bs4 import BeautifulSoup
from urllib.request import urlopen, Request
```

```python
all_teams = ['ATL','BRK','BOS','CHO','CHI','CLE','DAL','DEN','DET','GSW','HOU','IND','LAC','LAL','MEM','MIA','MIL','MIN',
        'NOP','NYK','OKC','ORL','PHI','PHO','POR','SAC','SAS','TOR','UTA','WAS']
name, pos, teams = [], [], []
```

```python
def clean_name(name):
    tokens = name.split(',')
    return tokens[1] + ' ' + tokens[0]
```

```python
def scrape_positions(team):
    res = requests.get('https://www.basketball-reference.com/teams/{0}/2020.html'.format(team))
    soup = BeautifulSoup(res.text, 'lxml')
    table = soup.find('table', attrs={'id':'roster'})
    for tr in table.find_all('tr'):
        name_td = tr.find('td', attrs={'data-stat':'player'})
        pos_td = tr.find('td', attrs={'data-stat':'pos'})
        if name_td is not None:
            name.append(clean_name(name_td.get('csk')))
            pos.append(pos_td.text)
            teams.append(team)
```

```python
for team in all_teams:
    scrape_positions(team)
```

```python
df = pd.DataFrame({'player name':name, 'team':teams, 'position':pos})
df.to_csv('./../OutputCSVs/all_player_positions.csv', header=True, index=False)
```
